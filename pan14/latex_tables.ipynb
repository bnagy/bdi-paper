```python
import pandas as pd
```

```python
all_cols = [
    "Tests",
    "U_{count}",
    "High Confidence",
    "U_{low}",
    "U_{hi}",
    "FP",
    "Prec",
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "PAN21",
    "PAN14",
    "classifier",
    "Corpus",
]
show_cols = [
    "Corpus",
    "Tests",
    "U_{count}",
    "U_{low}",
    "U_{hi}",
    "Prec",
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "PAN21",
    "PAN14",
]
highlight_cols = [
    "Prec",
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "PAN21",
    "PAN14",
]
```

```python
eval_results = pd.read_csv("eval_results.csv", index_col=0)
eval_results.columns = all_cols
df = eval_results.classifier.str.split("+", expand=True)
df.columns = ["Verifier", "Vectorizer", "Shifter"]
eval_results = eval_results.join(df)
eval_results["Vectorizer"].replace(
    to_replace={"vectorizer_24": "2,3,4-grams", "vectorizer_25": "2,3,4,5-grams"},
    regex=True,
    inplace=True,
)
eval_results["Corpus"].replace({"_": "\\_"}, regex=True, inplace=True)
eval_results
```

```python
eval_results.loc[0]
```

```python
k25 = eval_results.query("classifier=='Kestemont GI, Minmax + vectorizer_25 + fitted'")
```

```python
s = k25[show_cols].style.hide(axis="index").format(precision=3)

print(s.to_latex(hrules=True))
```

```python
bm25 = eval_results.query("classifier=='BDI, Minmax + vectorizer_25 + manual'")
```

```python
s = bm25[show_cols].style.hide(axis="index").format(precision=3)

print(s.to_latex(hrules=True))
```

```python
all_micro_cols = [
    "classifier",
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "PAN21",
    "Prec",
    "PAN14",
]
show_micro_cols = [
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "Prec",
    "PAN21",
    "PAN14",
    "PAN14 (unranked)",
]
highlight_micro_cols = [
    "Prec",
    "AUC",
    "C@1",
    "F_{05u}",
    "F_1",
    "Brier",
    "PAN21",
    "PAN14",
    "PAN14 (unranked)",
]
```

```python
micro_results = pd.read_csv("global_micro.csv", index_col=0)
micro_results.columns = all_micro_cols
micro_results
```

```python
df = micro_results.classifier.str.split("+", expand=True)
df.columns = ["Verifier", "Vectorizer", "Shifter"]
micro_results = micro_results.join(df)
```

```python
micro_results
```

```python
micro_results["Vectorizer"].replace(
    to_replace={"vectorizer_24": "2,3,4-grams", "vectorizer_25": "2,3,4,5-grams"},
    regex=True,
    inplace=True,
)
```

```python
unranked_micro_results = pd.read_csv("unranked_micro.csv", index_col=0)
unranked_micro_results.columns = all_micro_cols
df = unranked_micro_results.classifier.str.split("+", expand=True)
df.columns = ["Verifier", "Vectorizer", "Shifter"]
unranked_micro_results = unranked_micro_results.join(df)
unranked_micro_results["Vectorizer"].replace(
    to_replace={"vectorizer_24": "2,3,4-grams", "vectorizer_25": "2,3,4,5-grams"},
    regex=True,
    inplace=True,
)
unranked_micro_results
```

```python
micro_results["PAN14 (unranked)"] = unranked_micro_results["PAN14"]
```

```python
micro_results
```

```python
s = (
    micro_results.groupby(["Verifier", "Vectorizer", "Shifter"])[show_micro_cols]
    .agg("sum")
    .style.highlight_max(props="textbf:--rwrap;", subset=highlight_micro_cols)
    .format(precision=3)
)

print(s.to_latex(hrules=True))
```
